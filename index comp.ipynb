```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read in the S&P 500 constituents csv file and clean the data
sp_df = pd.read_csv('sp500.csv')

#Make column names more readable
sp_df.columns = ['RIC', 'Company_Name', 'market_cap', 'net_debt', 'ev', 'PE_FY0', 'PE_FY1', 'PE_FY2', 'PCF_FY0', 'PCF_FY1', 'PCF_FY2', 'PRev_FY0', 'PRev_FY1', 'PRev_FY2', 'PBV_FY0', 'PBV_FY1', 'PBV_FY2', 'EV_EBITDA_FY0', 'EV_EBITDA_FY1', 'EV_EBITDA_FY2', 'EV_Rev_FY0', 'EV_Rev_FY1', 'EV_Rev_FY2']

# Clense columns of commas and convert to float
sp_df['market_cap'] = sp_df['market_cap'].str.replace(',', '').astype(float)
sp_df['ev'] = sp_df['ev'].str.replace(',', '').astype(float)
sp_df['net_debt'] = sp_df['net_debt'].str.replace(',', '').astype(float)
# Convert all other columns to float
sp_df.iloc[:, 2:] = sp_df.iloc[:, 2:].apply(pd.to_numeric, errors='coerce')

#Sort by market cap
sp_df.sort_values("market_cap", ascending=False, inplace=True)

#Show the top 10 companies by market cap
sp_df.head(10)
```

```python
# Do the same for the Russell 1000

r_df = pd.read_csv('russell1000.csv')

r_df.columns = ['RIC', 'Company_Name', 'market_cap', 'net_debt', 'ev', 'PE_FY0', 'PE_FY1', 'PE_FY2', 'PCF_FY0', 'PCF_FY1', 'PCF_FY2', 'PRev_FY0', 'PRev_FY1', 'PRev_FY2', 'PBV_FY0', 'PBV_FY1', 'PBV_FY2', 'EV_EBITDA_FY0', 'EV_EBITDA_FY1', 'EV_EBITDA_FY2', 'EV_Rev_FY0', 'EV_Rev_FY1', 'EV_Rev_FY2']

r_df['market_cap'] = r_df['market_cap'].str.replace(',', '').astype(float)
r_df['ev'] = r_df['ev'].str.replace(',', '').astype(float)
r_df['net_debt'] = r_df['net_debt'].str.replace(',', '').astype(float)

r_df.iloc[:, 2:] = r_df.iloc[:, 2:].apply(pd.to_numeric, errors='coerce')
r_df.sort_values("market_cap", ascending=False, inplace=True)

r_df.head(10)
```

```python
# Mark the S&P 500 constituents in the Russell 1000
r_df["in_sp"] = r_df["RIC"].isin(sp_df["RIC"])

# Create a subset of the Russell 1000 that is not in the S&P 500
r_df_subset = r_df[r_df["in_sp"] == False]


```

```python
# Get the market cap of the 50% perncentile of the S&P
market_cap = sp_df["market_cap"].quantile(0.5)
```

```python
# Show the of companies in the Russell 1000 that are greater than the 30% percentile of the S&P 500
r_df_subset[ r_df["market_cap"] > market_cap]  
```
